#### Drop the duplicate values & drop null values

```python
df_copy.duplicated().sum()
df_copy.drop_duplicates(inplace=True)
df_copy.dropna(inplace=True)
df_copy.shape[0]
data = df_copy.copy()
data.shape[0]
```

```python
data["datetime"] = pd.to_datetime(data["last_update"])
data["day"] = data["datetime"].dt.day
data["month"] = data["datetime"].dt.month
data["year"] = data["datetime"].dt.year

data["hour"] = data["datetime"].dt.hour
data = data.drop(columns=["last_update", "station", "country", "datetime"])
# data = data.pop(columns=['last_update'])
data.head()
```

#### Encode Categorical columns 

It is used to convert categorical text to numerical values. It assign a unique integers to each distinct category. Typically ranging from 0 to n-1.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

data["city"] = encoder.fit_transform(data["city"])
data["state"] = encoder.fit_transform(data["state"])
data["pollutant_id"] = encoder.fit_transform(data["pollutant_id"])

data.head()
```

#### Select Features

```python
features = [
    'city',
    'state',
    'pollutant_id',
    'year',
    'month',
    'day',
    'hour',
]

traget = 'pollutant_avg'

X = data[features]
y = data[traget]

data.to_csv("../Dataset/processed_data.csv", index=False)
```
